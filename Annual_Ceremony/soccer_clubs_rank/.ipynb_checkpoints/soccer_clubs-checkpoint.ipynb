```python
def bar_club() -> Bar:
    background_color_js = (
        "new echarts.graphic.LinearGradient(0, 0, 0, 1, "
        "[{offset: 0, color: '#c86589'}, {offset: 1, color: '#06a7ff'}], false)"
    )
   
    c = (
        Bar(init_opts=opts.InitOpts(
                bg_color={
                    "type": "pattern",
                    "image": JsCode("img"),
                    "repeat": "no-repeat",
                },
            height='1000px',
            # width='1800px'
            ))
        .add_xaxis(club['club_name'].values.tolist()[:20][::-1])
        .add_yaxis("", club['point'].values.tolist()[:20][::-1], label_opts=opts.LabelOpts(position='inside', formatter="{b}: {c}"),)
        .reversal_axis()
        .set_global_opts(title_opts=opts.TitleOpts(title="2019俱乐部排行", title_textstyle_opts=opts.TextStyleOpts(color="white")),
                        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-30),
                                                axisline_opts=opts.AxisLineOpts(is_show=False),),
                         yaxis_opts=opts.AxisOpts(
                            axislabel_opts=opts.LabelOpts(formatter=JsCode ("""
                                                            function (value) {
                                                            for(x in value){
                                                            console.log(mydata[value])

                                                            }
                                                            return '{' + value + '|}';
                                                        }"""),
                                                         rich=rich
                                                         ))
                        )
        .set_series_opts(itemstyle_opts={
            "normal": {
                "color": JsCode("""new echarts.graphic.LinearGradient(0, 0, 0, 1, [{
                    offset: 0,
                    color: '#FF6461'
                }, {
                    offset: 0.5,
                    color: '#FFB1B0'
                },
                {offset: 1, color: '#F6CB90'}], )"""),
                
            }},)
    )
    c.add_js_funcs(
        """
        var img = new Image(); img.src = 'file:///C:/Work/code/Python/data_combat/Annual_Ceremony/soccer_clubs_rank/bk8.jpg';
        var mydata={'Liverpool_FC': 'file:///C:/Work/code/Python/data_combat/Annual_Ceremony/soccer_clubs_rank/pic/Liverpool.png', 'Manchester City': 2, 'Barcelona': 3};
        """
    )
    return c
```

```python
def pie_rosetype(data) -> Pie:
    c = (
        Pie(init_opts=opts.InitOpts(
                bg_color={
                    "type": "pattern",
                    "image": JsCode("img"),
                    "repeat": "no-repeat",
                },
            height='1000px',
            # width='1800px'
            ))
        .add(
            "",
            data,
            radius=["30%", "75%"],
            center=["50%", "50%"],
            rosetype="radius",
            label_opts=opts.LabelOpts(formatter="{b}: {c}"),
        )
        .set_global_opts(title_opts=opts.TitleOpts(title=""),
                        )
    )
    c.add_js_funcs(
        """
        var img = new Image(); img.src = 'file:///C:/Work/code/Python/data_combat/Annual_Ceremony/soccer_clubs_rank/bk8.jpg';
        """
    )
    return c
```

```python
def line_top7() -> Line:
    c = (
        Line()
        .add_xaxis(date.values.tolist()[:-1])
        .add_yaxis("Liverpool FC", Liverpool_FC.values.tolist()[:-1])
#         .add_yaxis("Liverpool FC rank", Liverpool_FC_rank.values.tolist()[:-1])
        .add_yaxis("Manchester City", Manchester_City.values.tolist()[:-1])
        .add_yaxis("Barcelona", Barcelona.values.tolist()[:-1])
        .add_yaxis("Bayern München", Bayern_München.values.tolist()[:-1])
        .add_yaxis("Juventus", Juventus.values.tolist()[:-1])
        .add_yaxis("Paris Saint-Germain", Paris_Saint_Germain.values.tolist()[:-1])
        .add_yaxis("Real Madrid", Real_Madrid.values.tolist()[:-1])
        .set_global_opts(title_opts=opts.TitleOpts(title=""))
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),
            markpoint_opts=opts.MarkPointOpts(
                data=[
                    opts.MarkPointItem(type_="max", name="最大值"),
                    opts.MarkPointItem(type_="min", name="最小值"),
                ]
            )
        )
    )
    return c
```
