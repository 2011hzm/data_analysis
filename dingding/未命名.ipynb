```python
import pandas as pd
```

```python
df = pd.read_csv('appstore_data1.csv', sep='|')
```

```python
df
```

```python
df_new = df.drop_duplicates(subset=['username', 'title'])
```

```python
def set_score(d):
    score = d.split('（')[0]
    return score
```

```python
df_new['score'] = df_new['score'].apply(set_score)
```
